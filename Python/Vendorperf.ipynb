```python
import pandas as pd

df = pd.read_excel("vendor_superstore.xlsx", sheet_name="Sheet1")

df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df['Selling Price'] = df['Selling Price'].astype(float)
df['Cost Price'] = df['Cost Price'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df['Profit'] = df['Selling Price'] - df['Cost Price']
```

```python
df['Sales'] = df['Selling Price'] * df['Quantity']
```

```python
df.describe()
```

```python
top_vendors = df.groupby("Vendor Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_vendors
```

```python
top_vendors.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Vendors by Sales")
plt.xlabel("Vendor Name")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_vendors = df.groupby("Vendor Name")["Sales"].sum().sort_values(ascending=True).head(10)

top_vendors

```

```python
top_vendors.plot(kind="bar", figsize=(10,5))

plt.title("Bottom 10 Vendors by Sales")
plt.xlabel("Vendor Name")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Create Profit column (if not already)
df['Profit'] = df['Selling Price'] - df['Cost Price']

# Top 10 vendors by profit
top_vendors = df.groupby("Vendor Name")["Profit"].sum().sort_values(ascending=False).head(10)

print(top_vendors)
```

```python
plt.figure()
top_vendors.plot(kind='bar')

# Labels & Title
plt.title("Top 10 Vendors by Profit")
plt.xlabel("Vendor Name")
plt.ylabel("Total Profit")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Lead Time'] = (df['Ship Date'] - df['Order Date']).dt.days
top_slow_vendors = df.groupby("Vendor Name")["Lead Time"].mean().sort_values(ascending=True).head(10)

print(top_slow_vendors)
```

```python
df['Sales'] = df['Selling Price'] * df['Quantity']
sales_by_region = df.groupby("Supplier Region")["Sales"].sum().sort_values(ascending=True)

print(sales_by_region)
```

```python
plt.figure()
sales_by_region.plot(kind='bar')

plt.title("Total Sales by Supplier Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix (Easy View)")
plt.show()
```

```python
for col in corr.columns:
    for row in corr.index:
        if col != row:
            value = corr.loc[row, col]
            if value > 0.7:
                print(f"{row} and {col} are Highly Positively Correlated ({value:.2f})")
            elif value < -0.7:
                print(f"{row} and {col} are Highly Negatively Correlated ({value:.2f})")
```

```python
corr_pairs = corr.unstack()
corr_pairs = corr_pairs.sort_values(ascending=False)

print(corr_pairs.drop_duplicates().head(10))
```
